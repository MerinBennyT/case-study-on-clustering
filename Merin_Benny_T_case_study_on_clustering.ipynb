```python
for col in df.columns:
  if df[col].dtype!='object':
    q1=df[col].quantile(0.25)
    q2=df[col].quantile(0.75)

    iqr=q2-q1
    upper=q2+1.5*iqr
    lower=q1-1.5*iqr
    df[col]=np.where(df[col]>upper,upper,df[col])
    df[col]=np.where(df[col]<lower,lower,df[col])
```

# Kmeans

```python
from sklearn.cluster import KMeans
```

```python

WCSS=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(df)
  WCSS.append(kmeans.inertia_)
```

```python
WCSS
```

```python
plt.plot(range(1,11),WCSS,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## **here the optimal non of cluster is 3**

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42,n_init='auto')
y_pred=kmeans.fit_predict(df)
```

```python
y_pred
```

```python
plt.scatter(df.loc[y_pred == 0, df.columns[0]], df.loc[y_pred == 0, df.columns[1]], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(df.loc[y_pred == 1, df.columns[0]], df.loc[y_pred == 1, df.columns[1]], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(df.loc[y_pred == 2, df.columns[0]], df.loc[y_pred == 2, df.columns[1]], s = 100, c = 'green', label = 'Cluster 3')

plt.title('Clusters')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
